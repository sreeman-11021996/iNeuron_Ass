## 1. Handle Missing Values

```python
# 1. Handle Missing Values
features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>0]
for feature in features_nan:
    print("The feature is {} and missing values is {}%".format(feature,np.round(dataset[feature].isnull().mean()*100,2)))
```

#### a. categorical features

```python
## Replacing categorical features
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
```

```python
categorical_features
```

```python
# 1st technique - fill with "Missing"

def replace_cat_features(dataset,categorical_features):
    data = dataset.copy()
    data[categorical_features] = data[categorical_features].fillna("Missing")
    return data
```

```python
replace_cat_features(dataset,categorical_features)
```

```python
# 1st technique - fill with Mode

def replace_mode_cat_features(dataset,categorical_features):
    data = dataset.copy()
    data[categorical_features] = data[categorical_features].fillna(data[categorical_features].mode())
    return data
```

```python
replace_mode_cat_features(dataset,categorical_features)
```

#### b. numerical features:

```python
numerical_features_nan = [feature for feature in dataset.columns if dataset[feature].dtypes!="O" and 
                         dataset[feature].isnull().sum()>0]
```

```python
numerical_features_nan
```

```python
## since there are outliers we are going to replace with median
for feature in numerical_features_nan:
    median_values = dataset[feature].median()
    ## create a new feature to capture nan values
    dataset[feature+"_nan"] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_values,inplace=True)
```

#### c. Temporal Variables (date time vaiables)

```python
for feature in ["YearBuilt","YearRemodAdd","GarageYrBlt"]:
        dataset[feature+"_age_of_house"] = dataset["YrSold"] - dataset[feature]
```
